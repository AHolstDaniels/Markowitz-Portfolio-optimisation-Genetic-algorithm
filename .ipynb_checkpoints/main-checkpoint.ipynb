```python
import statistics
from scipy.stats import norm
#Mean 
def cal_mean_var(arr):
    Return = []

    for i in range(1,len(arr)):
        Return.append((arr[i]-arr[i-1])/arr[i-1])
    Return.pop(0)
    mu = statistics.mean(Return)

    sigma = np.std(Return)

    return mu, sigma 
```

```python
Mu = []
Sigma = []
for i in range(1,20):
    print(i)
    for j in range(1,20):
        for k in range(1,20):
            mu , sigma = cal_mean_var(SPX_Close*i/(i+j+k) + TSX_Close*j/(i+j+k) + STOX_Close*k/(i+j+k))
            Mu.append(mu)
            Sigma.append(sigma)
```

```python
Mu1 = np.array(Mu)/max(abs(np.array(Mu)))
Sigma1 = np.array(Sigma)/max(Sigma)
for i in range(len(Mu)):
    if Sigma1[i] < 0.5:
        Sigma1.append(Sigma[i])
        Mu1.append(Mu[i])
```

```python
SPX_Close     = SPX['Close/Last'][::-1]   /float(SPX['Close/Last'][::-1][:1])   -1
SPX_Close = SPX_Close[45:]
TSX_Close    = TSX['Close/Last'][::-1]      /float(TSX['Close/Last'][::-1][:1])     -1
TSX_Close = TSX_Close[41:]
STOX_Close    = STOX['Close'][::-1]       /float(STOX['Close'][::-1][:1])     -1
STOX_Close = STOX_Close[43:]
SSE_Close     = SSE['Close'][::-1]       /float(SSE['Close'][::-1][:1])     -1
```

```python
plt.figure(figsize=(16,6))
plt.plot(dates, SPX_Close *100, label="S&P 500")
plt.plot(dates, TSX_Close *100, label="TSX")
plt.plot(dates, STOX_Close*100, label="STOXX")
plt.plot(dates, SSE_Close*100, label="SSE")
plt.xlabel('Time')
plt.ylabel('Cumulative Return (in %)')
plt.margins(x=0.005,y=0.02)
plt.axhline(y=0, xmin=0, xmax=1, linestyle='--', color='k')
plt.legend()
plt.show()
```

```python
BuyHold_EquiWeight = SPX_Close*0.25 + TSX_Close*0.25 + STOX_Close*0.25 + SSE_Close*0.25
plt.figure(figsize=(16,6))
plt.plot(dates, BuyHold_EquiWeight*100, label="Buy and Hold all with equi weights")
plt.xlabel('Time')
plt.ylabel('Cumulative Return (in %)')
plt.margins(x=0.005,y=0.02)
plt.axhline(y=0, xmin=0, xmax=1, linestyle='--', color='k')
plt.legend()
plt.show()
```
